```python
# Read Data file and Create data frame

df = pd.read_csv('customers.csv', sep="\t")
display(df)
```

```python
# Display the first 5 rows and the last 3 rows 
print("First 5 rows:")
display(df.head())
print("\nLast 3 rows:")
display(df.tail(3))
```

```python
# Show quick statistics of your data
df.describe()
```

```python
# Show the data type of each column
print("\nData types of each column:")
print(df.dtypes)
```

```python
# Showing the number of columns and rows in the data
num_rows, num_columns = df.shape
print(f"\nNumber of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# Showing the list of columns in the data frame
print("\nList of columns:")
print(df.columns.tolist())
```

```python
# Showing the number of duplicated rows in the data
num_of_duplicates_rows = df.duplicated().sum()
print(f"\nNumber of duplicated rows: {num_of_duplicates_rows}")
```

```python
# 8. Rename the columns belonging to Products by removing Mnt from them
df.rename(columns=lambda x: x.replace('Mnt', '').replace('Prods', '').replace('Products', '') if ('Mnt' in x) or ('Prods' in x) or ('Products' in x) else x, inplace=True)
df.rename(columns={'Wines': 'Wines', 'Fruits': 'Fruits', 'Meat': 'Meat', 'Fish': 'Fish', 'Sweet': 'Sweet', 'Gold': 'Gold'}, inplace=True)
print(df.columns.tolist())
```

```python
# 9. Rename the columns belonging to Place by removing Num and Purchases from them
df.rename(columns=lambda x: x.replace('Num', '').replace('Purchases', '') if ('Num' in x) or ('Purchases' in x) else x, inplace=True)
df.rename(columns={'NumWeb': 'Web', 'NumCatalog': 'Catalog', 'NumStorePurchases': 'Store'}, inplace=True)
print(df.columns.tolist())
```

```python
# 10. Keep only NumDealsPurchases column in the Promotion section
columns_to_keep = ['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome', 'Dt_Customer', 'Recency', 'Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold', 'Deals', 'Web', 'Catalog', 'Store', 'WebVisitsMonth', 'Complain', 'Z_CostContact', 'Z_Revenue']
df = df[columns_to_keep]
print(df.columns.tolist())
```

```python
# 11. Delete the column NumWebVisitsMonth
# Note Num has been removed from NumWebVisitsMonth in the last code 
df.drop(columns=['WebVisitsMonth'], inplace=True)
print(df.columns.tolist())
```

```python
# 12. Create a new derived column named Age
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']
display(df.head())
```

```python
# 13. Create a new column named Spending that holds the total amount spent on all product categories
df['Spending'] = df[['Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold']].sum(axis=1)
display(df.head())
```

```python
# 14. Change the Dt_Customer type from object to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
print(df.dtypes)
```

```python
# 15. Show the unique values of the column Marital_Status
print(df['Marital_Status'].unique())
```

```python
# 16. Classify customers into two segments for marital status
df['Marital_Status'].replace(['Divorced', 'Single', 'Absurd', 'Widow'], 'Alone', inplace=True)
df['Marital_Status'].replace(['Married', 'Together'], 'Not Alone', inplace=True)
```

```python
display(df['Marital_Status'])
```

```python
# 17. Show the unique values of the column Education
print(df['Education'].unique())
```

```python
# 18. Change the column Education values
df['Education'].replace(['Basic', '2n Cycle'], 'Undergraduate', inplace=True)
df['Education'].replace(['Master', 'PhD', 'Graduation'], 'Postgraduate', inplace=True)
print(df['Education'].unique())
```

```python
# 19. Investigate the presence of outliers in the Income column 
# Outliers are the values that are far from the other data points

print(df['Income'])
```

```python
# 19. Investigate the presence of outliers in the Income column 
# Outliers are the values that are far from the other data points

#Q1 = First quartile
Q1 = df['Income'].quantile(0.25)

#Q1 = Third quartile
Q3 = df['Income'].quantile(0.75)

# IQR = interquartile range
IQR = Q3 - Q1

# Removing the outlier values from income column
df = df[~((df['Income'] < (Q1 - 1.5 * IQR)) | (df['Income'] > (Q3 + 1.5 * IQR)))]
print(df['Income'])
```

```python
# 20. Show quick statistics for the column Income
print(df['Income'].describe())
```

```python
# 21. Show the number of missing values for each column
print(df.isnull().sum())
```

```python
# 22. Fill the missing values with the average

# From the last Query we know that only income column has missing values

# Filling missing values in Income columns with their respective means

df['Income'].fillna(df['Income'].mean(), inplace=True)
display(df)
```

```python
# 23. Create a new column named Children to hold the total number of children for every customer

df['Children'] = df['Kidhome'] + df['Teenhome']
display(df)
```

```python
# 24. Create a new column Has_Complaint
df['Has_Complaint'] = df['Complain'].apply(lambda x: 'Has complaint' if x == 1 else 'No complaint')
display(df)
```

```python
# 25. Show the first 5 rows of customers with ‘No complaint’
no_complaint_customers = df[df['Has_Complaint'] == 'No complaint'].head(5)
display(no_complaint_customers)
```

```python
# Create a new DataFrame with selected columns
selected_columns = ['Age', 'Spending', 'Marital_Status', 'Education', 'Income', 'Children', 'Dt_Customer', 'Has_Complaint']
new_df = df[selected_columns].copy()
display(new_df)
```

```python
# Create a histogram for the number of children per customer
plt.hist(new_df['Children'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Number of Children')
plt.ylabel('Frequency')
plt.title('Number of Children per Customer')
plt.show()
```

```python
# Histogram for Income Distributionn of Customers
plt.hist(new_df['Income'], bins=20, color='green', edgecolor='black')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution of Customers')
plt.show()
```

```python
# Boxplot for Age Distribution
sns.boxplot(y=new_df['Age'], color='lightgoldenrodyellow')
plt.ylabel('Age')
plt.title('Boxplot of Age')
plt.show()
```
