```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
from IPython.display import display

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots that we will create

# Read the data

df = pd.read_csv(r'F:\CanProfPersonal\Projects\Project1\movies.csv')
# print(df.head())

#to show data in a table
display(df.head())


```

```python
# Let's find any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# dropping the rows where there are any null or nan values
df = df.dropna()
```

```python

# Now Let's find any missing data percent

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# checking the data types

df.dtypes
```

```python
# Changing the data types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# gross and budget column data types changes to int and it just looks better let's see
df.head()
```

```python
# The year column and the year in the release date are not same 
# so we are going to create new column to have a year as of release date
# so for this we just take first four values of release date column and put 
# that value to year column

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

display(df.head())
```

```python
#order by gross revenue

df.sort_values(by=['gross'], inplace=False, ascending=False)

# Setting the dataframe

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# From the above table we get to know that 

#  There is high corelation between budget and gross i.e of 0.740247

# so to visualize it better lets use heatmap
```

```python
correlation_matrix = df.corr(method = 'pearson',numeric_only = True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correaltion Matrix for Numeric Features')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
# What we have done below is 
# changed data type of some columns to numeric
# so that we can find a corelation between the columns
# it just represent a particular string with a particular number

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
# df_numerized
# to show only few rows using head
df_numerized.head()
        
```

```python
# Here is the coorelation between all the columns
correlation_matrix = df_numerized.corr(method = 'pearson',numeric_only = True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correaltion Matrix for Numeric Features')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```

```python
df_numerized.corr(method = 'pearson',numeric_only = True)
```

```python
correlation_mat = df_numerized.corr(method = 'pearson',numeric_only = True)

corr_pairs = correlation_mat.unstack()

corr_pairs

```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
