```python
# Step 2: Display the first 5 rows and last 3 rows
print("First 5 rows:")
print(df.head(5))

print("\nLast 3 rows:")
print(df.tail(3))
```

```python
print("\nQuick statistics:")
print(df.describe())
```

```python
print("\nData types of each column:")
print(df.dtypes)
```

```python
# Step 5: Show the number of columns and rows
num_rows, num_cols = df.shape
print(f"\nNumber of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
```

```python
# Step 6: Show the list of columns
print("\nList of columns:")
print(df.columns)
```

```python
# Part II: Data Preparation and Cleaning
# Step 7: Show the number of missing values for each column
missing_values = df.isnull().sum()
print("\nNumber of missing values for each column:")
print(missing_values)
```

```python
# Step 7: Show unique values in the column species_id
unique_species = df['species_id'].unique()
print("\nUnique values in the column 'species_id':")
print(unique_species)

```

```python
# Step 8 : Create a list of unique site IDs (plot_id) found in the data
site_names = df['plot_id'].unique()
print("\nUnique site IDs (plot_id):")
print(site_names)
```

```python
# Step 8 (part a): Number of unique sites
num_unique_sites = len(site_names)
print(f"\nNumber of unique sites: {num_unique_sites}")
```

```python
# Step 8 (part b): Number of unique species
num_unique_species = df['species_id'].nunique()
print(f"Number of unique species: {num_unique_species}")
```

```python
# Step 8 (part c): Difference between len(site_names) and df['plot_id'].nunique()
difference = len(site_names) - df['plot_id'].nunique()
print(f"Difference between len(site_names) and df['plot_id'].nunique(): {difference}")
```

```python
# Step 9: Count the number of recorded individuals by gender
gender_count = df['sex'].value_counts()
print("\nNumber of recorded individuals by gender:")
print(gender_count)
```

```python
# Step 10: Summarize weight values for each site
weight_summary = df.groupby('plot_id')['weight'].describe()
print("\nSummary of weight values for each site:")
print(weight_summary)
```

```python
# Step 11: Count the number of samples for each species
species_count = df['species_id'].value_counts()
print("\nNumber of samples for each species:")
print(species_count)
```

```python
# Step 12: Multiply all weight values by 2
df['weight'] = df['weight'] * 2
print("\nWeight values multiplied by 2:")
print(df['weight'])
```

```python
#Part III: Data Visualization
import matplotlib.pyplot as plt
# Step 13: Ensure figures appear inline in Jupyter Notebook
%matplotlib inline
# Step 14: Create a quick bar chart to show the number of samples by species
species_count.plot(kind='bar', figsize=(8, 6))
plt.title('Number of Samples by Species')
plt.xlabel('Species')
plt.ylabel('Number of Samples')
plt.show()
```

```python
# Step 15: Create a plot of average weight across all species per site
average_weight_per_site = df.groupby('plot_id')['weight'].mean()
average_weight_per_site.plot(kind='bar', figsize=(10, 6))
plt.title('Average Weight Across Species per Site')
plt.xlabel('Site ID')
plt.ylabel('Average Weight')
plt.show()
```

```python
# Step 16: Create a plot of total males versus total females for the entire dataset
total_gender_count = df.groupby('sex').size()
total_gender_count.plot(kind='bar', figsize=(6, 6))
plt.title('Total Males vs Females')
plt.xlabel('Gender')
plt.ylabel('Total Count')
plt.show()
```

```python
# Step 17: Create a stacked bar plot showing total weight by sex for each site
stacked_data = df.groupby(['plot_id', 'sex'])['weight'].sum().unstack()
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Total Weight by Sex for Each Site')
plt.xlabel('Site ID')
plt.ylabel('Total Weight')
plt.legend(title='Sex')
plt.show()
```
